```python
import pandas as pd
```

```python
xl = pd.ExcelFile("./Voluntary-Registry-Offsets-Database--v9-December-2023.xlsx")
acr = pd.read_excel(xl, "ACR Retirements")
car = pd.read_excel(xl, "CAR Retirements")
gold = pd.read_excel(xl, "Gold Retirements")
ver = pd.read_excel(xl, "VCS Issuances & Retirements")
```

```python
acr
```

```python
_gold = gold[["ID", "Retirement Date", "Notes", "Quantity"]]
_gold = _gold.rename(columns={"Retirement Date": "date", "Notes": "notes", "ID": "id", "Quantity": "total"})
_gold["type"] = 0
_gold
```

```python
_ver = ver[["Project ID", "Retirement/ Cancellation Date", "Retirement Beneficiary", "Credits Retired"]]
_ver = _ver.rename(columns={"Retirement/ Cancellation Date": "date", "Retirement Beneficiary": "notes", "Project ID": "id", "Credits Retired": "total"})
_ver["type"] = 1
_ver.dropna(inplace=True)
_ver
```

```python
_acr = acr[["Project ID", "Status Effective", "Retirement Reason Details", "Quantity of Offset Credits"]]
_acr = _acr.rename(columns={"Status Effective": "date", "Retirement Reason Details": "notes", "Project ID": "id", "Quantity of Offset Credits": "total"})
_acr["type"] = 2
_acr
```

```python
import json

all = pd.concat([_gold, _ver, _acr])
all.dropna(inplace=True)
all.to_csv("data/all_sales.csv", index=False)

all["date"] = all["date"].astype(str)
values = all.values.tolist()
headers = ["id", "date", "notes", "total", "type"]

with open("data/all_sales.json", "w") as outfile:
    json.dump({'headers': headers, 'values': values}, outfile)
# all.to_json("data/all_sales.json", orient="records")
```
